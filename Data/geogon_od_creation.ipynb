```python
import pandas as pd 
import geopandas as gpd
```

```python
# read in the polygon data (found from census)
gdf = gpd.read_file('cb_2018_us_county_20m.shp')
gdf['FIPS'] = gdf['STATEFP'] + gdf['COUNTYFP']
gdf = gdf[['FIPS', 'geometry']]
```

```python
# read in overdose df 
geo_od = pd.read_csv("geo_od2.csv")

# Convert to geopandas 
geo_od = gpd.GeoDataFrame(geo_od)

# drop the current geometry column bc its wrong 
geo_od = geo_od.drop('geometry', axis = 1)

# TURN FIPS COL TO STRING 
geo_od['FIPS'] = geo_od['FIPS'].astype(str).str.zfill(5)
geo_od
```

```python
# MERGE 
polygon = geo_od.merge(gdf, on = ['FIPS'], how = 'left')
polygon 
```
