# Adding geographical polygon geometry to od_full

```python
import os
import pandas as pd
pd.set_option('display.max_columns', None) 
import geopandas as gpd 
```

```python
os.chdir('C:\\Users\\rhian\\Downloads\\Capstone\\drugs\\Data')

od = pd.read_csv('od_full.csv')

# Convert FIPS to string 
od['FIPS'] = od['FIPS'].astype(str).str.zfill(5) 
od
```

```python
# Load the json file with county coordinates
geoData = gpd.read_file('https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/US-counties.geojson')
geoData = geoData[['id', 'geometry']]
geoData.columns = ['FIPS', 'Geometry']
geoData
```

```python
# merging together 
od = od.merge(geoData, on = ['FIPS'], how = 'left')
od
```
